## Read Data from JSON

```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

## Add New Item into Inventory

```python
def addnewitem():
    prod_id = str(input("Enter product id:"))
    if prod_id not in records.keys():
        name = str(input("Enter name: "))
        pr = int(input("Enter price: "))
        qn = int(input("Enter quantity:"))
        Exp=str(input("Enter expire date: "))
        net_amount=str(input("Enter net amount: "))
        records[prod_id] = {'name': name,'Exp': Exp,'Net amount':net_amount,'qn': qn, 'pr': pr}
        js = json.dumps(records)
        fd = open("record.json",'w')
        fd.write(js)
        fd.close()
        another=input("Do you want to add another product?(Y/N)")
        if another=="Y" or another=="y" or another =="Yes" or another=="yes":
            return(addnewitem())
        else:
            print("Successfully you added new product/products in our inventory")
            pass
    else:
        addin=input("This product is available in our inventory. Do you add more quantities of this product?(Y/N)")
        if addin == "Y" or addin=="y" or addin =="Yes" or addin=="yes":
            qn = int(input("Enter quantity:"))
            records[prod_id]["qn"]=int(records[prod_id]["qn"])+qn
            js = json.dumps(records)
            fd = open("record.json",'w')
            fd.write(js)
            fd.close()
            another=input("Do you want to add another product?(Y/N)")
            if another=="Y" or another=="y" or another =="Yes" or another=="yes":
                return(addnewitem())
            else:
                print("Successfully you added new product/products in our inventory")
                pass
        else:
            another=input("Do you want to add another product?(Y/N)")
            if another=="Y" or another=="y" or another =="Yes" or another=="yes":
                return(addnewitem())
            else:
                print("You didn't add any product in our inventory.")
                pass
            
            
addnewitem()
```
